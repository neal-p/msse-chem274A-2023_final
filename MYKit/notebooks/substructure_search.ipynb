```python
from pathlib import Path
import sys

sys.path.insert(0, str(Path('..')))
from MYKit import Mol, SDFToMol
```

```python
# Search function
def search(query, sdf_root, maxresults=20):
    res = 0
    for sdf in sdf_root.glob("*sdf"):
        other = SDFToMol(sdf)
        if other.hasSubstructMatch(query):
            res += 1
            if res > maxresults:
                break
            print("FOUND SUBSTRUCTURE!", sdf)
            display(other)
            yield(sdf)
```

```python
# Set SDF dir path
sdf_root = Path("../../sdf")
```

```python
# A very unique query
query_sdf = sdf_root / "NDCUAPJVLWFHHB-UHNVWZDZSA-N.sdf"
query = SDFToMol(query_sdf)
query
```

```python
matches = list(search(query, sdf_root))
```
